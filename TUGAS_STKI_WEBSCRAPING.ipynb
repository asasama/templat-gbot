```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

3. Setelah berhasil mengimpor library BeautifulSoup dan Request, Kami melakukan pemanggilan pada website target



```python
alamat = "https://iprice.co.id/kamera-foto/dslr/"
req = Request(alamat, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'})
```

4. Setelah halaman web berhasil dibuka, jalankan Beautiful Soup untuk melakukan ekstraksi tabel yang akan kita ambil. Ekstraksi dilakukan pada Tag div dan class = 'pk cN lQ kF vI vA vS vW vH vL vM vN pH vC'.

```python
html = urlopen(req).read()
data = BeautifulSoup(html, 'html.parser')

div = data.find_all('div', {'class' : 'pk cN lQ kF vI vA vS vW vH vL vM vN pH vC'})[0]

print(div)
```

5. Selanjutnya kita akan coba simpan masing-masing baris yang berisi informasi lengkap dari sebuah produk (Nama Produk, Harga, Rating, Deskripsi dll) kedalam sebuah list bernama rows. Informasi ini terdapat pada Tag HTML a (link) dengan 'class':'iy oT cN ei kL br-

'.

```python
rows = div.findAll('a', href = True, attrs = {'class': 'iy oT cN ei kL br-8'})
products = []
prices = []
product_types = []

for row in rows:
  name = row.find('h3', attrs = {'class':'u7 vV ow-wb bu ve R e4 fw6 tcc vR'})
  price = row.find('div', attrs = {'class' : 'db gM ei b vV be hE vB ff ellipsis-1'})
  product_type = row.find('span', attrs = {'class' : 'fw6 u6 ga n'})

  products.append(name.text)
  prices.append(price.text)
  product_types.append(product_type.text)

products
```
